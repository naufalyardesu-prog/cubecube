```python
class order:
    def __init__(self,order_id,customer_name,order_date,total_amount):
        self.order_id=order_id
        self.customer_name=customer_name
        self.order_date=order_date
        self.total_amount=total_amount

    #fungsi hitung pajak
    def tax_amount(self,tax_rate):
        tax_amount = int(self.total_amount*tax_rate)
        return tax_amount

    #fungsi display
    def display_order(self):
        print(f"ID : {self.order_id}")
        print(f"NAMA : {self.customer_name}")
        print(f"TOTAL : {self.total_amount}")
        print(f"PAJAK : {self.tax_amount(0.1)}")


class OrderProcessor:
    def __init__(self):
        self.orders=[]

    #fungsi menambah pesanan
    def add_order(self, order):
        self.orders.append(order)
        print(f'pesanan ditambahkan: {order.order_id}')

    #fungsi hitung total pendapatan
    def calculate_total_revenue(self):
        total = 0
        for order in self.orders:
            total += order.total_amount
        return total

    #fungsi hitung total pajak
    def calculate_total_tax(self, tax_rate):
        total_tax = 0
        for order in self.orders:
            total_tax += order.tax_amount(tax_rate)
        return total_tax

    #fungsi display prosesor
    def display_prosesor(self):
        print(f'TOTAL PENDAPATAN: {self.calculate_total_revenue()}')
        print(f'TOTAL PAJAK: {self.calculate_total_tax(0.1)}')


pesanan1 = order("001", "Budi", "2024-01-01",100000)
pesanan2 = order("002", "Dono", "2024-12-12",250000)
pesanan3 = order("002", "Lala", "2024-12-12",250000)
prosesor = OrderProcessor()
print('=====================DISPLAY PESANAN====================')
order.display_order(pesanan1)
order.display_order(pesanan2)
order.display_order(pesanan3)
print('====================PROSESOR PESANAN====================')
prosesor.add_order(pesanan1)
prosesor.add_order(pesanan2)
prosesor.add_order(pesanan3)
prosesor.display_prosesor()

```
